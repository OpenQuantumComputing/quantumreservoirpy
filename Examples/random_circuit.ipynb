# Random_circuit

Utilities include the function 'random_circuit', which creates a random circuit of a given depth, number of qubits and a set of gates. The function is a modification of qiskits own implementation, and therefore use the same set of operators. 

The complete set of standard gates can be found at https://qiskit.org/documentation/apidoc/circuit_library.html

```python
from qreservoirpy import utilities
from qiskit import QuantumCircuit
from qiskit.circuit.library.standard_gates import IGate, XGate, ZGate, CXGate, YGate, HGate
```

## Creating a random circuit

```python
circ = utilities.random_circuit(4, gates=[
    IGate, XGate, YGate, ZGate, CXGate, HGate
], depth=4)

circ.draw('mpl')
```

## Creating a random gate/instruction

One can, of course, having obtained the random circuit, make a gate/instruction using to_gate/to_instruction. 

```python
gate = circ.to_gate(label="RC")

qc = QuantumCircuit(6)

qc.append(gate, range(4))
qc.append(gate, range(2, 6))
qc.draw('mpl')
```
